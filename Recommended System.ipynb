```python
import pandas as pd
import ast
```

```python
movies = pd.read_csv("C:/Users/shahn/Downloads/archive/tmdb_5000_movies.csv")
credits = pd.read_csv("C:/Users/shahn/Downloads/archive/tmdb_5000_credits.csv")
movies.head(3)
```

```python
credits.head(3)
```

```python
movies = movies.merge(credits,on="title")
movies.head(3)
```

```python
movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
movies.dtypes
```

```python
movies.isnull().sum()
```

```python
movies.dropna(subset=['overview'],inplace=True)
```

```python
type(movies['genres'][0])
```

```python
def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name'])
    return L

movies['genres'] = movies['genres'].apply(convert)
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies['genres'].iloc[0]
```

```python
def convert_cast(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter<3:
            L.append(i['name'])
            counter += 1
        else:
            break
    return L

movies['cast'] = movies['cast'].apply(convert_cast)       
```

```python
movies['cast'].iloc[0]
```

```python
def fetch_directer(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L
movies['crew'] = movies['crew'].apply(fetch_directer)
```

```python
movies['crew'].iloc[0]
```

```python
movies['overview'] = movies['overview'].apply(lambda x: x.split())
```

```python
movies['genres'] = movies['genres'].apply(lambda x: [i.replace(" ", "") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x: [i.replace(" ", "") for i in x])
movies['cast'] = movies['cast'].apply(lambda x: [i.replace(" ", "") for i in x])
movies['crew'] = movies['crew'].apply(lambda x: [i.replace(" ", "") for i in x])
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
df = movies[['movie_id','title','tags']]
```

```python
df['tags'] = df['tags'].apply(lambda x: " ".join(x))
```

```python
df['tags'] = df['tags'].apply(lambda x: x.lower())
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
vectors=cv.fit_transform(df['tags']).toarray()
```

```python
vectors.shape
```

```python
import nltk
from nltk.stem.porter import PorterStemmer

ps = PorterStemmer()

def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)

df['tags'] = df["tags"].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
vectors=cv.fit_transform(df['tags']).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity

similarity = cosine_similarity(vectors)
```

```python
similarity.shape
```

```python
def recommend(movie):
    movie_index = df[df['title'] == movie].index[0]
    distances = similarity[movie_index]

    movie_list = sorted(
        enumerate(distances),
        key=lambda x: x[1],
        reverse=True
    )[1:6]

    for i in movie_list:
        print(df.iloc[i[0]].title)
```

```python
recommend('Avatar')
```
